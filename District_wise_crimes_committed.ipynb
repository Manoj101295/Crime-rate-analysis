```python
# since there "TOTAL " named district is there, i will replace this value with specific state, i assume  this data is not 
# specified by which district is but able to understand which state is.
distict_name=list(df['DISTRICT'].values)  
state_name=list(df['STATE/UT'].values)
for i in range(len(distict_name)):
    if distict_name[i] =='TOTAL':
        distict_name[i]=(state_name[i] +' TOTAL')
```

```python
#Since there is no much data about CUSTODIAL RAPE, so we will forward with RAPE columns only because 
#RAPE='CUSTODIAL RAPE'  + 'OTHER RAPE
# also KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS + KIDNAPPING AND ABDUCTION OF OTHERS =KIDNAPPING & ABDUCTION
#THEFT=AUTO THEFT+OTHER THEFT
```

```python
data=df[['STATE/UT','Distics', 'YEAR', 'MURDER', 'ATTEMPT TO MURDER',
       'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER', 'RAPE','KIDNAPPING & ABDUCTION',
        'DACOITY', 'PREPARATION AND ASSEMBLY FOR DACOITY', 'ROBBERY', 'BURGLARY', 'THEFT',
        'RIOTS', 'CRIMINAL BREACH OF TRUST',
       'CHEATING', 'COUNTERFIETING', 'ARSON', 'HURT/GREVIOUS HURT',
       'DOWRY DEATHS', 'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY',
       'INSULT TO MODESTY OF WOMEN', 'CRUELTY BY HUSBAND OR HIS RELATIVES',
       'IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES',
       'CAUSING DEATH BY NEGLIGENCE', 'OTHER IPC CRIMES', 'TOTAL IPC CRIMES',
       ]]
```

```python
data.shape
```

```python
##we will now analyze the crime counts


df1=data.copy()
df1.drop(['STATE/UT','Distics','YEAR','TOTAL IPC CRIMES','OTHER IPC CRIMES'],axis=1, inplace=True)

columns = list(df1.sum().sort_values(ascending = False).index)

toltal_number_of_indivisual_crimes = df1.sum()
td=toltal_number_of_indivisual_crimes[columns]
td
```

```python
# lets plot the bar plot number of crimes vs crimes
plt.figure(figsize=(12,6))
plt.title("TOTAL CRIMES")
plt.ylabel('CRIMES');
plt.xlabel('NUMBER OF CRIMES IN MILLIANS');
sns.barplot(td,td.index)
plt.plot()
```

```python
# lets analyze the total IPC crime over the year,state,district
```

```python
ta=df.groupby("YEAR")['TOTAL IPC CRIMES'].sum()
ta
```

```python
# lets plot the graph total crrime per year
plt.figure(figsize=(12,6))
plt.title("TOTAL CRIMES PER YEAR")
plt.ylabel('TOTAL CRIMES IN MILLIANS');
plt.xlabel('YEARS');
ta.plot()
plt.plot()
```

```python
#lets plot the graph total crimes  per year
plt.figure(figsize=(12,6))
sns.barplot(x =ta.index, y =ta ).set(title = "total crimes in Millians per year")
plt.plot()
```

```python
#from graph we can see that total crime is incresing as year
# lets plot the pie chart for better view
plt.figure(figsize=(10,10))
plt.pie(ta,labels = ta.index)
plt.show()
```

```python
# the  below graphs are  used to describe how indivisual crime count is vary year

for feature in columns:
    plt.figure(figsize=(10,6))
    data.groupby('YEAR')[feature].sum().plot()
    plt.xlabel('Year')
    plt.ylabel(feature)
    plt.show()
```

```python
## lets anlyze the total count per state
tb = df.groupby("STATE/UT")["TOTAL IPC CRIMES"].sum().sort_values(ascending = False)
tb
```

```python
## lets analyze it by plotting the bar plot
plt.figure(figsize=(20,15))
sns.barplot(x =tb,y =tb.index ).set(title = "total crimes Milians per State")
plt.plot()
```

```python
# from below graphs we can analyze the number of crimes per state
for feature in columns:
    plt.figure(figsize=(15,8))
    data.groupby("STATE/UT")[feature].sum().sort_values(ascending = False).plot.bar()
    plt.xlabel("STATE/UT")
    plt.ylabel(feature)
    plt.show()
```

```python
# lets analyze it through districs
tc = df.groupby("Distics")["TOTAL IPC CRIMES"].sum().sort_values(ascending = False)
tc[:50]
```

```python
## Displaying the top 15 districts
plt.figure(figsize=(25,15))
sns.barplot(x =tc[:15], y =tc[:15].index ).set(title = "total crimes for top 15 districts")
plt.plot()
```
